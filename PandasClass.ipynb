```python
def decide_tax_percentage(sal):
    if sal>=200000:
        return 20
    elif sal>=150000:
        return 15
    elif sal>=100000:
        return 10
    else:
        return 5
```

```python
exc['tax%'] = exc.Salary.apply(decide_tax_percentage)
exc
```

```python
exc['tax%'] = exc.Salary.apply(lambda x: 20 if x>=200000 else 15 if x>=150000 else 10 if x>=100000 else 5)
exc
```

```python
exc['tax%'] = exc.Salary.apply(lambda x: x*.20 if x>=200000 else x*.15 if x>=150000 else x*.10 if x>=100000 else x*.5)
exc
```

```python
# First, make sure salary and tax columns have proper data types
exc['Salary'] = pd.to_numeric(exc['Salary'], errors='coerce')
exc['tax%'] = pd.to_numeric(exc['tax%'], errors='coerce')

# Now calculate net salary after tax
exc['Net_Salary'] = exc['Salary'] - (exc['Salary'] * exc['tax%'] / 100)
exc['Net_Salary'] = exc['Salary'] - (exc['Salary'] * exc['tax%'] / 100)
exc
```
